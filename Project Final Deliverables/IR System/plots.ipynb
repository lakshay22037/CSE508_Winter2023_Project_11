```python
import matplotlib.pyplot as plt
```

```python
# Time vs Dimensionality

time = [7.86797,7.38243,8.2025,8.43399,7.8349]
dimensions = [1254,1280,1434,1612,1792]

plt.scatter(dimensions, time,c=['red','blue','red','red','blue'])
plt.plot(dimensions, time,c='black')
plt.title('Embedding Size vs Average Query Procesing Time')
plt.xlabel('Embedding Dimension')
plt.ylabel('Average Time (ms)')
plt.show()
```

```python
# Space vs Dimensionality

size = [482,492,550,619,688]
dimensions = [1254,1280,1434,1612,1792]

plt.scatter(dimensions, size,c=['red','blue','red','red','blue'])
plt.plot(dimensions, size,c='black')
plt.title('Embedding Size vs Inverted Index Size')
plt.xlabel('Embedding Dimension')
plt.ylabel('Index Size (megabytes)')
plt.show()
```

```python
# MAP on seen data
map = [0.9258,0.944,0.9254,0.9209,0.9707]
dimensions = [1254,1280,1434,1612,1792]

plt.scatter(dimensions, map ,c=['red','blue','red','red','blue'])
plt.plot(dimensions, map ,c='black')
plt.title('Embedding Size vs MAP (Seen Data)')
plt.xlabel('Embedding Dimension')
plt.ylabel('Mean Average Precision')
plt.show()
```
